```python
import perceptron as perc
import img_to_gif
import numpy as np
import matplotlib.pyplot as plt
```

## The Perceptron algorithm for linearly seperable data 

```python
# We first load the data 

#X,y = load_data_set(show = False, xy = True)
X = np.load('./data/X_lin_sep.npy')
y = np.load('./data/y_lin_sep.npy')


fig, ax = plt.subplots(nrows=1, ncols=1, dpi=500, figsize=(8,6))
perc.plot_data_points(ax, X, y)


plt.savefig('output/lin_sep/figure.png')
plt.show()
```

```python
X_train, X_test, y_train, y_test = perc.train_test_split(X, y, train_size=0.75)

perc.perceptron_anim(X, y, X_train, y_train, X_test, y_test, 'output/lin_sep/anim', iteration_level=False, p=1, n_iter=2)

perceptron = perc.Perceptron()
perceptron.fit(X_train, y_train,idx=1)

perceptron.score(X_test, y_test)
```

```python
img_to_gif.images_to_gif('output/lin_sep/anim', './perceptron1.gif' )
```
